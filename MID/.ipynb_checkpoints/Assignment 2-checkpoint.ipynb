# Multi-Class Neural Network

This assignment implements a three-hidden-layer neural network from scratch for a synthetic five-class classification problem. A dataset with five classes is generated, the network is trained on this data, and its performance is evaluated using standard classification metrics and visualizations.


```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import make_blobs
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelBinarizer
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix, roc_auc_score
import seaborn as sns
```

## Dataset and Task

The goal is to create a synthetic dataset with five distinct classes and use it to train and test a neural network with three hidden layers. The dataset is generated with `make_classification`, producing eight input features per sample that form a mix of informative and non-trivial decision boundaries, and the labels indicate which of the five classes each sample belongs to.


```python
from sklearn.datasets import make_classification

n_samples = 3000
n_features = 8
n_classes = 5

X, y = make_classification(
    n_samples=n_samples,
    n_features=n_features,
    n_informative=5,
    n_redundant=0,
    n_repeated=0,
    n_classes=n_classes,
    n_clusters_per_class=1,
    class_sep=2.0,
    random_state=42,
)
```

```python
X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.2, stratify=y, random_state=42
)
```

```python
lb = LabelBinarizer()
y_train_bin = lb.fit_transform(y_train)
y_test_bin = lb.transform(y_test)

if y_train_bin.shape[1] == 1:
    y_train_bin = np.hstack([1 - y_train_bin, y_train_bin])
```

```python
def relu(z):
    return np.maximum(0.0, z)

def relu_derivative(z):
    return (z > 0).astype(np.float32)

def softmax(z):
    z_stable = z - np.max(z, axis=1, keepdims=True)
    exp_z = np.exp(z_stable)
    return exp_z / np.sum(exp_z, axis=1, keepdims=True)
```

## Network Design and Modification

The original single-output network is extended to handle five output classes by using three hidden layers and an output layer with five neurons. The forward pass now produces a probability distribution over the five classes using softmax, and the loss, gradients, and backpropagation steps are adapted to use a multi-class cross-entropy objective.

Each hidden layer computes

$$
z^{(l)} = a^{(l-1)} W^{(l)} + b^{(l)}, \quad a^{(l)} = \max(0, z^{(l)})
$$

and the output layer computes

$$
z^{(\text{out})} = a^{(L-1)} W^{(L)} + b^{(L)}, \quad p = \text{softmax}(z^{(\text{out})}).
$$

The training objective is the multi-class cross-entropy:

$$
L = -\frac{1}{N} \sum_{n=1}^N \sum_{c=1}^C y_{n,c} \log p_{n,c}.
$$



```python
class NeuralNetwork:
    def __init__(self, input_dim, hidden_dims, output_dim, lr=0.01):
        self.lr = lr
        self.sizes = [input_dim] + hidden_dims + [output_dim]
        L = len(self.sizes) - 1
        self.W = [np.random.randn(self.sizes[i], self.sizes[i+1]) * np.sqrt(2.0 / self.sizes[i])
                  for i in range(L)]
        self.b = [np.zeros((1, self.sizes[i+1]), dtype=np.float32) for i in range(L)]

    def forward(self, X):
        a = X.astype(np.float32)
        acts = [a]      
        pre_acts = []   
        for i in range(len(self.W) - 1):
            z = np.dot(a, self.W[i]) + self.b[i] #z = ax + b
            pre_acts.append(z)
            a = relu(z)
            acts.append(a)
        z_out = np.dot(a, self.W[-1]) + self.b[-1]
        pre_acts.append(z_out)
        probs = softmax(z_out)
        acts.append(probs)
        return pre_acts, acts

    def compute_loss(self, probs, y_onehot):
        eps = 1e-12
        N = y_onehot.shape[0]
        return -np.sum(y_onehot * np.log(probs + eps)) / N

    def backward(self, pre_acts, acts, X_batch, y_batch_onehot):
        N = X_batch.shape[0]
        L = len(self.W)
        grads_W = [None] * L
        grads_b = [None] * L

        probs = acts[-1]
        delta = (probs - y_batch_onehot) / N 
        grads_W[-1] = np.dot(acts[-2].T, delta)
        grads_b[-1] = np.sum(delta, axis=0, keepdims=True)

        downstream = delta
        for i in range(L-2, -1, -1):
            z = pre_acts[i]
            d_relu = relu_derivative(z)
            downstream = np.dot(downstream, self.W[i+1].T) * d_relu
            grads_W[i] = np.dot(acts[i].T, downstream)
            grads_b[i] = np.sum(downstream, axis=0, keepdims=True)

        for i in range(L):
            self.W[i] -= self.lr * grads_W[i]
            self.b[i] -= self.lr * grads_b[i]

    def predict(self, X):
        _, acts = self.forward(X)
        return np.argmax(acts[-1], axis=1)

    def predict_proba(self, X):
        _, acts = self.forward(X)
        return acts[-1]
```

```python
input_dim = n_features
hidden_dims = [64, 32, 16]
output_dim = n_classes
```

```python
model = NeuralNetwork(input_dim, hidden_dims, output_dim, lr=0.01)
```

```python
epochs = 120
batch_size = 64
n_train = X_train.shape[0]
steps_per_epoch = int(np.ceil(n_train / batch_size))

train_losses = []
val_losses = []
train_accs = []
val_accs = []
```

The gradient at the output layer uses the softmax + cross-entropy simplification:

$$
\frac{\partial L}{\partial z^{(\text{out})}} = \frac{1}{N} \left(p - y\right),
$$

which is then propagated backward through each layer using the chain rule and the derivative of ReLU:

$$
\frac{\mathrm{d}}{\mathrm{d}z} \max(0, z) =
\begin{cases}
1 & z > 0 \\
0 & z \le 0
\end{cases}
$$


```python
for epoch in range(epochs):
    perm = np.random.permutation(n_train)
    X_shuf = X_train[perm]
    y_shuf = y_train[perm]
    
    y_shuf_onehot = np.zeros((n_train, output_dim))
    y_shuf_onehot[np.arange(n_train), y_shuf] = 1

    epoch_losses = []
    for step in range(steps_per_epoch):
        start = step * batch_size
        end = min(start + batch_size, n_train)
        X_batch = X_shuf[start:end]
        y_batch_onehot = y_shuf_onehot[start:end]

        pre, acts = model.forward(X_batch)
        loss = model.compute_loss(acts[-1], y_batch_onehot)
        epoch_losses.append(loss)

        model.backward(pre, acts, X_batch, y_batch_onehot)

    # metrics
    train_losses.append(np.mean(epoch_losses))
    y_train_pred = model.predict(X_train)
    train_accs.append(accuracy_score(y_train, y_train_pred))

    pre_val, acts_val = model.forward(X_test)
    val_loss = model.compute_loss(acts_val[-1], (np.eye(output_dim)[y_test]).astype(np.float32))
    val_losses.append(val_loss)
    y_val_pred = np.argmax(acts_val[-1], axis=1)
    val_accs.append(accuracy_score(y_test, y_val_pred))

    if (epoch + 1) % 10 == 0 or epoch == 0:
        print(f"Epoch {epoch+1}/{epochs} - train_loss: {train_losses[-1]:.4f}, "
              f"train_acc: {train_accs[-1]:.4f}, val_loss: {val_loss:.4f}, val_acc: {val_accs[-1]:.4f}")
```

## Challenges and Implementation Notes

During implementation, several practical points had to be handled carefully:

- **Label representation:** The data are integer class labels in $\{0,\dots,4\}$, but the cross-entropy loss expects one-hot vectors. This was addressed by constructing one-hot encodings for training batches and for the validation set.
- **Numerical stability of softmax and log:** To avoid overflow/underflow when computing softmax, the maximum logit per sample is subtracted before exponentiation. A small constant $\varepsilon$ is also added inside the logarithm in the loss to avoid $\log(0)$.
- **Weight initialization:** He initialization is used for all layers with ReLU activations. This keeps the variance of activations more stable across layers and helps the network train reliably.
- **Gradient flow through multiple layers:** Backpropagation was extended to three hidden layers by iteratively propagating the downstream gradient through each weight matrix and ReLU derivative. This ensures every layer receives an appropriate gradient update.


```python
y_test_pred = model.predict(X_test)
print("\nFinal Test Accuracy:", accuracy_score(y_test, y_test_pred))
print("\nClassification Report:\n", classification_report(y_test, y_test_pred))
```

## Results and Observations

The final test accuracy is around 0.96, indicating that the network correctly classifies the majority of samples in this 5-class problem. The per-class precision, recall, and F1-scores are all relatively high, although there are small differences between classes.

The confusion matrix shows very strong performance, with most predictions lying on the diagonal and only a few misclassifications between neighboring classes. 


```python
cm = confusion_matrix(y_test, y_test_pred)
plt.figure(figsize=(6,5))
sns.heatmap(cm, annot=True, fmt='d', cmap='Blues', cbar=False)
plt.xlabel("Predicted")
plt.ylabel("Actual")
plt.title("Confusion Matrix")
plt.show()
```

```python
y_score = model.predict_proba(X_test)


auc_macro = roc_auc_score(y_test_bin, y_score, average='macro', multi_class='ovr')
print("Macro ROC AUC (OVR):", auc_macro)


plt.figure(figsize=(8, 6))
for i in range(output_dim):
    fpr, tpr, _ = roc_curve(y_test_bin[:, i], y_score[:, i])
    auc_i = auc(fpr, tpr)
    plt.plot(fpr, tpr, label=f"Class {i} (AUC = {auc_i:.2f})")

plt.plot([0, 1], [0, 1], 'k--')  # Diagonal line
plt.xlabel("False Positive Rate")
plt.ylabel("True Positive Rate")
plt.title("ROC Curves (One-vs-Rest)")
plt.legend(loc="lower right")
plt.show()
```

```python
plt.figure(figsize=(6,4))
plt.plot(range(1, epochs+1), train_losses, marker='o', label='train loss')
plt.plot(range(1, epochs+1), val_losses, marker='s', label='val loss')
plt.xlabel('Epoch')
plt.ylabel('Loss')
plt.title('Loss per Epoch')
plt.grid(True)
plt.legend()
plt.show()

plt.figure(figsize=(6,4))
plt.plot(range(1, epochs+1), train_accs, marker='o', label='train acc')
plt.plot(range(1, epochs+1), val_accs, marker='s', label='val acc')
plt.xlabel('Epoch')
plt.ylabel('Accuracy')
plt.title('Accuracy per Epoch')
plt.grid(True)
plt.legend()
plt.show()
```
